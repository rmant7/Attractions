```python
failed_list_of_cities = []

def get_place_pics(city,ammount_of_pic_to_get,perspective_type,size_of_image,we_can_save_image):
 
  retry_delay = 5 
  chep_trip_api = ''
  atraction_name = ''

  view_type = perspective_type
  

  print(f'we save attractions for city {city}')
  #read from json on server s3 
  whole_cities = read_from_aws_s3_json()
  
  #WE GET ALL ATTR IN THIS CITY
  for attraction_name in  whole_cities[city]:
    #define the way where we save iamge in base
    wehere_we_save = f'Landmark_of_the_cities/{city}/attractiions_images/{attraction_name}.jpg'
    
    try:
      resp = openai.Image.create(
        prompt= f"{view_type} {attraction_name} in {city}",
        n=ammount_of_pic_to_get,
        size=size_of_image
      
      )
      print(f'we got the bytes of the image {city} {attraction_name}')
      for each_resp in resp["data"]:
      
        img_url = each_resp['url']
        response = requests.get(img_url)
        #cont image bytes
        cont = response.content
        #FIERBASE SAVE
        if(we_can_save_image == 'fierbase'):
          print(f'now we try to upload {attraction_name} in city {city} to fierbase')
          
          #geting the app from base
          app_we_get = firebase_admin.get_app(name='Storage1')
          #setting the bucket from admin 
          bucket = storage.bucket(app=app_we_get)
          
          #set where to save image
          
          blob = bucket.blob(wehere_we_save)
          
          #upload from bytes
          blob.upload_from_string(cont, content_type='image/jpeg')
          
          #upload_to_fierbase_from_bytes(cont,wehere_we_save)
          print(f'we uploaded to fierbase')
        #LOCAL FOLDER SAVE  
        elif(we_can_save_image =='local'):
          print(f'we save localy for {city} attraction name {attraction_name}')
          with open(f'./Landmark_of_the_cities/{city}/{attraction_name}.jpg', 'wb') as f:
            f.write(cont)
          f.close()
        
    except Exception as e:
        if(city not in failed_list_of_cities):
          failed_list_of_cities.append(city)
        print(f"error {type(e)} on city name {city} attraction name {attraction_name} occurred: {str(e)}")
        time.sleep(retry_delay)
        pass
    
```
